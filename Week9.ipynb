**Import beberapa library berikut yang akan digunakan selama uji coba praktikum**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/PCVK_MM/Images/j.png')
plt.imshow(img)
```

**8. Buatlah program implementasi operasi closing dengan Structuring Element berbentuk square 3 x 3 tanpa dan dengan menggunakan library morphology dari openCV, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar j.png)**

```python
# bo 8 tanpa library
element = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
dilate_dst = cv2.dilate(img, element)
closing_dst = cv2.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Citra Hasil Closing')
```

```python
# nomer 8 dengan library

kernel = np.ones((3,3), np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing), plt.title('Citra Hasil Closing')
```

**9. Buatlah program implementasi operasi dilasi dengan Structuring Element berbentuk Cross 3 x 3, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar j.png)**

```python
# nomor 9 dengan library 3x3

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
cross3 = cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Dilasi 3x3')
```

**10. Buatlah program implementasi operasi dilasi dengan Structuring Element berbentuk Circular 5 x 5, sehingga menghasilkan tampilan seperti di bawah ini: (Gunakan gambar j.png)**
